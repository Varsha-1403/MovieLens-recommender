```python
import pandas as pd
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
from collections import defaultdict
from surprise.model_selection import train_test_split
import random
```

```python
# Load the datasets
movies = pd.read_csv('C:/source/Orion_Innovation_internship_repos/MovieLens-recommender/Dataset/movie.csv')
ratings = pd.read_csv('C:/source/Orion_Innovation_internship_repos/MovieLens-recommender/Dataset/rating.csv')

# Merge datasets on movieId
data = pd.merge(ratings, movies, on='movieId')

# Use only relevant columns for the recommendation system
data = data[['userId', 'movieId', 'rating']]

# Define the format for the data
reader = Reader(rating_scale=(0.5, 5.0))

# Load the data into the surprise library
dataset = Dataset.load_from_df(data, reader)

# Split the data into training and testing sets
trainset = dataset.build_full_trainset()

# Use SVD algorithm (Singular Value Decomposition)
algo = SVD()

# Train the algorithm on the trainset
algo.fit(trainset)

# To generate recommendations for a specific user
user_id = 1  # Change this to the userId for whom you want to make recommendations
user_ratings = data[data['userId'] == user_id]

# Get a list of all movie IDs
all_movie_ids = data['movieId'].unique()

# Get the movie IDs that the user has already rated
rated_movie_ids = user_ratings['movieId'].unique()

# Get the movie IDs that the user has not rated yet
unrated_movie_ids = [movie_id for movie_id in all_movie_ids if movie_id not in rated_movie_ids]

# Predict ratings for the unrated movies
predictions = [algo.predict(user_id, movie_id) for movie_id in unrated_movie_ids]

# Sort the predictions by estimated rating
predictions.sort(key=lambda x: x.est, reverse=True)

# Get the top 10 movie recommendations
top_10_recommendations = predictions[:10]

# Print the recommended movies
print("Top 10 movie recommendations for user {}: ".format(user_id))
for pred in top_10_recommendations:
    movie_id = pred.iid
    movie_title = movies[movies['movieId'] == movie_id]['title'].values[0]
    print(movie_title, pred.est)
```

```python

# Load the datasets
movies = pd.read_csv('C:/source/Orion_Innovation_internship_repos/MovieLens-recommender/Dataset/movie.csv')
ratings = pd.read_csv('C:/source/Orion_Innovation_internship_repos/MovieLens-recommender/Dataset/rating.csv')


# Merge datasets on movieId
data = pd.merge(ratings, movies, on='movieId')

# Use only relevant columns for the recommendation system
data = data[['userId', 'movieId', 'rating']]

# Define the format for the data
reader = Reader(rating_scale=(0.5, 5.0))

# Load the data into the surprise library
dataset = Dataset.load_from_df(data, reader)

# Split the data into training and testing sets
trainset, testset = train_test_split(dataset, test_size=0.2)

# Use SVD algorithm (Singular Value Decomposition)
algo = SVD()

# Train the algorithm on the trainset
algo.fit(trainset)

# Evaluate the model on the testset
predictions = algo.test(testset)

# Function to get top-N recommendations for all users
def get_top_n(predictions, n=10):
    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the n highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n

# Function to get top-N recommendations for a specific genre for random users
def get_top_n_for_genre(predictions, genre, movies, n=10):
    top_n = defaultdict(list)
    
    # Filter movies by genre (case insensitive and handling multiple genres)
    genre_movies = movies[movies['genres'].str.contains(genre, case=False, na=False)]
    genre_movie_ids = genre_movies['movieId'].values
    
    print(f"Found {len(genre_movie_ids)} movies for genre '{genre}'")
    
    # Map the predictions to each user
    for uid, iid, true_r, est, _ in predictions:
        if iid in genre_movie_ids:
            top_n[uid].append((iid, est))
    
    # Sort the predictions for each user and retrieve the n highest ones
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]
    
    return top_n

# Example usage
# Take genre as user input
input_genre = input("Enter the genre: ")

# Get the top 10 recommendations for the specified genre
top_n_genre = get_top_n_for_genre(predictions, input_genre, movies, n=10)

# Check if there are users with recommendations for the specified genre
if len(top_n_genre) > 0:
    # Select random users to display recommendations for
    num_users_to_display = min(5, len(top_n_genre))  # Ensure we don't sample more users than available
    random_users = random.sample(list(top_n_genre.keys()), num_users_to_display)

    # Print the recommended movies for the selected random users
    for uid in random_users:
        print(f"User {uid}:")
        for (iid, rating) in top_n_genre[uid]:
            movie_title = movies[movies['movieId'] == iid]['title'].values[0]
            print(f"  {movie_title}: {rating:.2f}")
else:
    print(f"No recommendations available for the genre '{input_genre}'")
```
